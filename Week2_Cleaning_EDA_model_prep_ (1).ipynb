## 2.3 Text Preprocessing Functions

```python
# Import necessary libraries
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy

# Download NLTK data if not already downloaded
# try:
#     stopwords.words('english')
# except LookupError:
#     nltk.download('stopwords')
# try:
#     WordNetLemmatizer().lemmatize('test')
# except LookupError:
#     nltk.download('wordnet')

# NOTE: nltk downloads are commented out because they are not needed if the original notebook setup handles them, or if the environment already has them.
# For safety, I'll include the necessary downloads here in case of a fresh environment.
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# Initialize NLP components
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
nlp = spacy.load('en_core_web_sm') # Already loaded in cell e9468048

def preprocess_text(text):
    # 1. Lowercasing
    text = str(text).lower()

    # 2. Remove URLs and emails
    text = re.sub(r'\S*@\S*|https?://\S+|www\.\S+', '', text)

    # 3. Remove special characters and numbers (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 4. Tokenization and Stopword Removal (using spaCy for tokenization for consistency with lemmatization)
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.text not in stop_words and token.is_alpha]

    # 5. Lemmatization (done within spaCy tokenization as token.lemma_)

    # Join tokens back into a single string
    return ' '.join(tokens)

# Test the preprocessing function with sample texts
print("[TEST] Testing preprocessing function:\n")

test_samples = [
    "Water pipe BURST on Main St! Emergency!! Call 911 NOW!!!",
    "Garbage not collected for 3 weeks at 123 Park Ave. Health hazard!",
    "Street light is out @ the corner of 5th & Broadway since Jan 2024"
]

for idx, sample in enumerate(test_samples, 1):
    print(f"Sample {idx}:")
    print(f"  Original: {sample}")
    try:
        cleaned = preprocess_text(sample)
        print(f"  Cleaned:  {cleaned}")
    except Exception as e:
        print(f"  Error: {str(e)}")
    print()

print("[SUCCESS] Preprocessing test complete!")
```

## 3.2 N-gram Frequency Analysis

```python
import collections
from nltk.util import ngrams

def get_top_ngrams(corpus, n=1, top=None):
    """
    Generate a list of top N-grams from a corpus.
    :param corpus: A pandas Series or list of text strings.
    :param n: The size of the n-gram (e.g., 1 for unigrams, 2 for bigrams).
    :param top: The number of top n-grams to return.
    :return: A list of (n-gram, frequency) tuples.
    """
    all_tokens = []
    for text in corpus:
        if isinstance(text, str):
            all_tokens.extend(text.split())

    n_grams_gen = ngrams(all_tokens, n)
    n_gram_counts = collections.Counter(n_grams_gen)
    return n_gram_counts.most_common(top)

# Extract unigrams (single words)
print("[STATS] Extracting top unigrams (single words)...\n")

top_unigrams = get_top_ngrams(df['text_cleaned'], n=1, top=20)

# Convert to dataframe for display
unigram_df = pd.DataFrame(top_unigrams, columns=['Word', 'Frequency'])
unigram_df['Word'] = unigram_df['Word'].apply(lambda x: x[0])

print("Top 20 Most Common Words:")
print(unigram_df.to_string(index=False))

# Visualize
plt.figure(figsize=(12, 6))
plt.barh(range(len(unigram_df)), unigram_df['Frequency'], color='skyblue', edgecolor='black')
plt.yticks(range(len(unigram_df)), unigram_df['Word'])
plt.xlabel('Frequency', fontsize=12)
plt.title('Top 20 Most Common Words in Complaints', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import nltk
nltk.download('punkt_tab')# Extract trigrams (three-word phrases)
print("[STATS] Extracting top trigrams (three-word phrases)...\n")

top_trigrams = get_top_ngrams(df['text_cleaned'], n=3, top=20)

# Convert to dataframe
trigram_df = pd.DataFrame(top_trigrams, columns=['Trigram', 'Frequency'])
trigram_df['Trigram'] = trigram_df['Trigram'].apply(lambda x: ' '.join(x))

print("Top 20 Most Common Trigrams:")
print(trigram_df.to_string(index=False))

# Visualize
plt.figure(figsize=(12, 6))
plt.barh(range(len(trigram_df)), trigram_df['Frequency'], color='lightcoral', edgecolor='black')
plt.yticks(range(len(trigram_df)), trigram_df['Trigram'])
plt.xlabel('Frequency', fontsize=12)
plt.title('Top 20 Most Common Trigrams in Complaints', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Analyze bigrams for each top category
print("[STATS] Analyzing bigrams for top 3 categories...\n")

top_3_categories = df['category'].value_counts().head(3).index

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for idx, category in enumerate(top_3_categories):
    # Get text for this category
    category_texts = df[df['category'] == category]['text_cleaned'].values

    # Get top bigrams
    category_bigrams = get_top_ngrams(category_texts, n=2, top=10)

    # Prepare data
    bigrams = [' '.join(bg[0]) for bg in category_bigrams]
    frequencies = [bg[1] for bg in category_bigrams]

    # Plot
    axes[idx].barh(range(len(bigrams)), frequencies, color='steelblue', edgecolor='black')
    axes[idx].set_yticks(range(len(bigrams)))
    axes[idx].set_yticklabels(bigrams)
    axes[idx].set_xlabel('Frequency', fontsize=10)
    axes[idx].set_title(f'{category}', fontsize=12, fontweight='bold')
    axes[idx].invert_yaxis()

plt.suptitle('Top 10 Bigrams by Category', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 4.1 Feature Engineering: TF-IDF

```python
# TF-IDF Vectorization
print("\n[STEP 1/7] Performing TF-IDF Vectorization...")

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_features=5000, # Limit to top 5000 features to manage dimensionality
    ngram_range=(1, 2) # Consider both unigrams and bigrams
)

# Fit and transform the cleaned text data
X_tfidf = tfidf_vectorizer.fit_transform(df['text_cleaned'])
y = df['category'] # Target variable

print("[SUCCESS] TF-IDF Vectorization complete.")
print(f"TF-IDF Matrix Shape: {X_tfidf.shape}")
print(f"Number of unique categories: {len(y.unique())}")
```

## 4.2 Feature Engineering: Word2Vec Embeddings

```python
# Word2Vec Embeddings
print("\n[STEP 2/7] Generating Word2Vec Embeddings...")

# Prepare text for Word2Vec (list of lists of words)
sentences = [text.split() for text in df['text_cleaned']]

# Train Word2Vec model
# Using a smaller vector size and window for this dataset's characteristics
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4, sg=0)

# Function to average word vectors for a given text
def get_word_embedding(tokens, model, vector_size):
    # Filter out words that are not in the model's vocabulary
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if vectors: # If there are any valid vectors, return their mean
        return np.mean(vectors, axis=0)
    else: # Otherwise, return a zero vector
        return np.zeros(vector_size)

# Apply the function to create document embeddings
X_w2v = np.array([get_word_embedding(s, w2v_model, 100) for s in sentences])

print("[SUCCESS] Word2Vec Embeddings generated.")
print(f"Word2Vec Embedding Matrix Shape: {X_w2v.shape}")
```

## 4.3 Data Splitting

```python
print("\n[STEP 3/7] Splitting data into training and testing sets...")

# Identify classes with only one member
class_counts = y.value_counts()
single_instance_classes = class_counts[class_counts == 1].index

if not single_instance_classes.empty:
    print(f"Removing {len(single_instance_classes)} classes with only one instance for stratified split.")
    # Create the boolean mask once
    mask = ~df['category'].isin(single_instance_classes)

    # Filter out rows corresponding to these classes from df and y
    df_filtered = df[mask]
    X_tfidf_filtered = X_tfidf[mask.values] # Use .values here to convert to numpy array
    X_w2v_filtered = X_w2v[mask.values]     # Use .values here to convert to numpy array
    y_filtered = y[mask] # For pandas Series, boolean indexing works directly

    print(f"Remaining data points: {len(df_filtered)}")
else:
    df_filtered = df
    X_tfidf_filtered = X_tfidf
    X_w2v_filtered = X_w2v
    y_filtered = y

# Split data for TF-IDF features
X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(
    X_tfidf_filtered, y_filtered, test_size=0.2, random_state=RANDOM_SEED, stratify=y_filtered
)

# Split data for Word2Vec features
X_train_w2v, X_test_w2v, y_train_w2v, y_test_w2v = train_test_split(
    X_w2v_filtered, y_filtered, test_size=0.2, random_state=RANDOM_SEED, stratify=y_filtered
)

print("[SUCCESS] Data splitting complete.")
print(f"TF-IDF Training set shape: {X_train_tfidf.shape}")
print(f"TF-IDF Testing set shape: {X_test_tfidf.shape}")
print(f"Word2Vec Training set shape: {X_train_w2v.shape}")
print(f"Word2Vec Testing set shape: {X_test_w2v.shape}")

# Check if y_train_tfidf and y_train_w2v are identical
if y_train_tfidf.equals(y_train_w2v) and y_test_tfidf.equals(y_test_w2v):
    print("Target variables for TF-IDF and Word2Vec splits are consistent.")
    # We can use a single y_train and y_test for both feature sets going forward
    y_train = y_train_tfidf
    y_test = y_test_tfidf
else:
    print("Warning: Target variable splits are not identical. Investigate.")
```

## 4.4 Baseline Model Training & Evaluation (TF-IDF Features)

```python
print("\n[STEP 4/7] Training and evaluating Logistic Regression with TF-IDF features...")

# Initialize and train Logistic Regression model
lr_tfidf = LogisticRegression(max_iter=1000, random_state=RANDOM_SEED, solver='saga', multi_class='multinomial')
lr_tfidf.fit(X_train_tfidf, y_train)

# Make predictions
y_pred_lr_tfidf = lr_tfidf.predict(X_test_tfidf)

# Evaluate the model
print("\n--- Logistic Regression (TF-IDF) Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr_tfidf):.4f}")
print(f"F1-Score (macro): {f1_score(y_test, y_pred_lr_tfidf, average='macro'):.4f}")
print(f"Classification Report:\n{classification_report(y_test, y_pred_lr_tfidf, zero_division=0)}")

# Cross-validation
print("\nPerforming 5-fold cross-validation...")
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)
cv_scores_lr_tfidf = cross_val_score(lr_tfidf, X_tfidf_filtered, y_filtered, cv=kf, scoring='f1_macro', n_jobs=-1)
print(f"Cross-validation F1-Score (macro): {cv_scores_lr_tfidf.mean():.4f} (+/- {cv_scores_lr_tfidf.std():.4f})")

print("\n[STEP 5/7] Training and evaluating Random Forest with TF-IDF features...")

# Initialize and train Random Forest model
rf_tfidf = RandomForestClassifier(n_estimators=100, random_state=RANDOM_SEED, n_jobs=-1)
rf_tfidf.fit(X_train_tfidf, y_train)

# Make predictions
y_pred_rf_tfidf = rf_tfidf.predict(X_test_tfidf)

# Evaluate the model
print("\n--- Random Forest (TF-IDF) Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf_tfidf):.4f}")
print(f"F1-Score (macro): {f1_score(y_test, y_pred_rf_tfidf, average='macro'):.4f}")
print(f"Classification Report:\n{classification_report(y_test, y_pred_rf_tfidf, zero_division=0)}")

# Cross-validation
print("\nPerforming 5-fold cross-validation...")
cv_scores_rf_tfidf = cross_val_score(rf_tfidf, X_tfidf_filtered, y_filtered, cv=kf, scoring='f1_macro', n_jobs=-1)
print(f"Cross-validation F1-Score (macro): {cv_scores_rf_tfidf.mean():.4f} (+/- {cv_scores_rf_tfidf.std():.4f})")
```

## 5.4 Advanced Feature Engineering: BERT Embeddings

```python
# Generate BERT Embeddings
print("\n[STEP 9/X] Generating BERT Embeddings...")

from transformers import AutoTokenizer, AutoModel
import torch

# Load a pre-trained BERT tokenizer and model
# 'bert-base-uncased' is a good general-purpose model
bert_model_name = 'bert-base-uncased'
bert_tokenizer = AutoTokenizer.from_pretrained(bert_model_name)
bert_model = AutoModel.from_pretrained(bert_model_name)

# Function to get BERT embeddings for a list of texts
def get_bert_embeddings(texts, tokenizer, model, max_length=128, batch_size=32):
    model.eval()
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]
        encoded_input = tokenizer(
            batch_texts,
            padding=True,
            truncation=True,
            max_length=max_length,
            return_tensors='pt'
        )
        with torch.no_grad():
            model_output = model(**encoded_input)
        # Use the [CLS] token embedding for sentence representation
        # You can also average all token embeddings or use other pooling strategies
        sentence_embeddings = model_output.last_hidden_state[:, 0, :].numpy()
        all_embeddings.append(sentence_embeddings)
    return np.vstack(all_embeddings)

# Generate BERT embeddings for the cleaned text
X_bert_embeddings = get_bert_embeddings(df_filtered['text_cleaned'].tolist(), bert_tokenizer, bert_model)

print("[SUCCESS] BERT Embeddings generated.")
print(f"BERT Embeddings Matrix Shape: {X_bert_embeddings.shape}")

# Prepare target variable (y) similar to previous steps
y_bert = y_filtered

# Split data for BERT features
X_train_bert, X_test_bert, y_train_bert, y_test_bert = train_test_split(
    X_bert_embeddings, y_bert, test_size=0.2, random_state=RANDOM_SEED, stratify=y_bert
)

print("[SUCCESS] Data splitting for BERT complete.")
print(f"BERT Training set shape: {X_train_bert.shape}")
print(f"BERT Testing set shape: {X_test_bert.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.model_selection import KFold, cross_val_score

RANDOM_SEED = 42 # Define RANDOM_SEED for this cell's scope

print("\n[STEP X/X] Training and evaluating Logistic Regression with BERT features...")

# Initialize and train Logistic Regression model
lr_bert = LogisticRegression(max_iter=1000, random_state=RANDOM_SEED, solver='saga', multi_class='multinomial')
lr_bert.fit(X_train_bert, y_train_bert)

# Make predictions
y_pred_lr_bert = lr_bert.predict(X_test_bert)

# Evaluate the model
print("\n--- Logistic Regression (BERT) Performance ---")
print(f"Accuracy: {accuracy_score(y_test_bert, y_pred_lr_bert):.4f}")
print(f"F1-Score (macro): {f1_score(y_test_bert, y_pred_lr_bert, average='macro'):.4f}")
print(f"Classification Report:\n{classification_report(y_test_bert, y_pred_lr_bert, zero_division=0)}")

# Cross-validation
print("\nPerforming 5-fold cross-validation...")
kf_bert = KFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED) # Using a new KFold for consistency with BERT splits
cv_scores_lr_bert = cross_val_score(lr_bert, X_bert_embeddings, y_bert, cv=kf_bert, scoring='f1_macro', n_jobs=-1)
print(f"Cross-validation F1-Score (macro): {cv_scores_lr_bert.mean():.4f} (+/- {cv_scores_lr_bert.std():.4f})")

print("\n[STEP X/X] Training and evaluating Random Forest with BERT features...")

# Initialize and train Random Forest model
rf_bert = RandomForestClassifier(n_estimators=100, random_state=RANDOM_SEED, n_jobs=-1)
rf_bert.fit(X_train_bert, y_train_bert)

# Make predictions
y_pred_rf_bert = rf_bert.predict(X_test_bert)

# Evaluate the model
print("\n--- Random Forest (BERT) Performance ---")
print(f"Accuracy: {accuracy_score(y_test_bert, y_pred_rf_bert):.4f}")
print(f"F1-Score (macro): {f1_score(y_test_bert, y_pred_rf_bert, average='macro'):.4f}")
print(f"Classification Report:\n{classification_report(y_test_bert, y_pred_rf_bert, zero_division=0)}")

# Cross-validation
print("\nPerforming 5-fold cross-validation...")
cv_scores_rf_bert = cross_val_score(rf_bert, X_bert_embeddings, y_bert, cv=kf_bert, scoring='f1_macro', n_jobs=-1)
print(f"Cross-validation F1-Score (macro): {cv_scores_rf_bert.mean():.4f} (+/- {cv_scores_rf_bert.std():.4f})")
```
